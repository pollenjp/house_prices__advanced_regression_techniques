```python
# データをtrain, testに分ける
# DataFrameのままsplit
df_train, df_test = \
    train_test_split(df, test_size=0.3, random_state=0)
```

```python
df_train.head()
```

```python
df_test.head()
```

```python
# X_df_train : 訓練データのパラメータ（変数）のDataFrame
X_df_train = df_train.iloc[:, :-1]  # 'SalePrice'以外のパラメータ
X_df_train.head()
```

```python
# y_df_test : 訓練データのtargetのDataFrame
y_df_train = df_train.iloc[:, -1]  # 'SalePrice'
y_df_train.head()
```

```python
# X_df_test : テストデータのパラメータ（変数）のDataFrame
X_df_test = df_test.iloc[:, :-1]  # 'SalePrice'以外のパラメータ
X_df_test.head()
```

```python
# y_df_test : テストデータのパラメータのDataFrame
y_df_test = df_test.iloc[:, -1]  # 'SalePrice'
y_df_test.head()
```

```python
# NaN値を取り除く(欠損値処理)
def remove_nan(dataframe):
    print("処理前：", dataframe.shape)
    dataframe = dataframe.dropna(axis=1)  # NaNを含む列を削除
    print("処理後：", dataframe.shape)
    return dataframe
```

```python
# 訓練データの欠損値処理
X_df_train = remove_nan(X_df_train)
X_df_train.head()
```

```python
# テストデータの欠損値処理
X_df_test = remove_nan(X_df_test)
X_df_test.head()
```

もし行を削除(dropna)していた場合は欠損値の発生が完全にランダム(MCR:Missing Completely at Random)でないと処理はデータ分布を大きく変えてしまうが、列を削除するのであればパラメータ数が減るだけなので、分布に影響は与えない（と思う）

### one-hot encodeing

```python
# train data
X_df_train = pd.get_dummies(X_df_train)
print(X_df_train.shape)
```

```python
# test data
X_df_test = pd.get_dummies(X_df_test)
print(X_df_test.shape)
```
